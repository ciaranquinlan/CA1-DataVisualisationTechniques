```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f500data= pd.read_csv("fortune500_data.csv")
```

```python
# Group the DataFrame by 'year' and find the leading company for each year
leading_companies = f500data.groupby('year').apply(lambda x: x.loc[x['revenue'].idxmax()])

# Create a plot of the leading company for each year
plt.figure(figsize=(10, 6))
plt.bar(leading_companies['year'], leading_companies['revenue'], color='blue', label='Revenue')
plt.bar(leading_companies['year'], leading_companies['profit'], color='green', label='Profit')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.title('Leading Company by Year')
plt.xticks(leading_companies['year'])

# Annotate the bars with company names
for i, row in leading_companies.iterrows():
    plt.annotate(row['company'], (row['year'], row['revenue']), xytext=(0, 10), textcoords='offset points', ha='center')

plt.legend(['Revenue', 'Profit'])
plt.tight_layout()
plt.show()
```

```python
# Function to plot the top 5 companies by total revenue for a given decade
def plot_top5_companies(start_year, end_year):
    # Filter the DataFrame for the specified decade
    decade_data = f500data[(f500data['year'] >= start_year) & (f500data['year'] <= end_year)]

    # Aggregate the data to find the top 5 companies by total revenue
    top5_companies = decade_data.groupby('company')['revenue'].sum().reset_index()
    top5_companies = top5_companies.sort_values(by='revenue', ascending=False).head(5)

    # Create a bar plot for the top 5 companies in the decade
    plt.figure(figsize=(10, 6))
    plt.barh(top5_companies['company'], top5_companies['revenue'], color='skyblue')
    plt.xlabel('Total Revenue')
    plt.ylabel('Company')
    plt.title(f'Top 5 Companies ({start_year}-{end_year}) by Total Revenue')
    plt.gca().invert_yaxis()  # Invert the y-axis for better visualization
    plt.show()

# Choose the decades you want to analyze (e.g., 1960s and 1970s)
plot_top5_companies(1950, 1959)
plot_top5_companies(1960, 1969)
plot_top5_companies(1970, 1979)
plot_top5_companies(1980, 1989)
plot_top5_companies(1990, 1999)
```

```python

# Function to get the top 5 companies by total revenue for a given decade
def top5_companies_by_decade(start_year, end_year):
    # Filter the DataFrame for the specified decade
    decade_data = f500data[(f500data['year'] >= start_year) & (f500data['year'] <= end_year)]

    # Aggregate the data to find the top 5 companies by total revenue
    top5_companies = decade_data.groupby('company')['revenue'].sum().reset_index()
    top5_companies = top5_companies.sort_values(by='revenue', ascending=False).head(5)

    return top5_companies

# Create a list of DataFrames, one for each decade
decades_dataframes = []

# Define the start and end years for the decades you want to analyze
decade_start_years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
for start_year in decade_start_years:
    end_year = start_year + 9  # Each decade spans 10 years
    top5_df = top5_companies_by_decade(start_year, end_year)
    decades_dataframes.append(top5_df)

# Concatenate the DataFrames into a single DataFrame
result_df = pd.concat(decades_dataframes, ignore_index=True)

# Display the result DataFrame
print(result_df)
```
